```python
def get_continious_categorical_data_columns(data):
    contnious_data_list = []
    categorical_data_list = []
    for col in data.columns:
        print(f"Col : {col} | Datatype : {data[col].dtypes}")
        if data[col].dtypes == "object":
            categorical_data_list.append(col)
        else:
            contnious_data_list.append(col)
    return categorical_data_list, contnious_data_list

categorical_data_list, contnious_data_list  = get_continious_categorical_data_columns(DATA_SET)
```

```python
Continues_Data_set=DATA_SET[contnious_data_list]
```

```python
Continues_Data_set = DATA_SET[contnious_data_list]

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for each column
Q1 = Continues_Data_set.quantile(0.25)
Q3 = Continues_Data_set.quantile(0.75)

# Calculate the IQR (Interquartile Range)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers: Keep only rows where data points are within the bounds
DATA_SET_cleaned = Continues_Data_set[~((Continues_Data_set < lower_bound) | (Continues_Data_set > upper_bound)).any(axis=1)]

print(f"Original data shape: {DATA_SET.shape}")
print(f"Data shape after removing outliers: {DATA_SET_cleaned.shape}")

# Find the rows that were removed
removed_rows_indices = DATA_SET[~DATA_SET.index.isin(DATA_SET_cleaned.index)].index

# Remove those rows from the original dataset
DATA_SET_cleaned_full = DATA_SET.drop(removed_rows_indices)

print(f"Updated original data shape after removing rows: {DATA_SET_cleaned_full.shape}")
```

```python
# Set up the figure for subplots
fig, axes = plt.subplots(nrows=len(merged_df['YrSold'].unique()), ncols=1, figsize=(12, 8), sharex=True)

# If there's only one subplot, axes will be a single object, not an array
if len(merged_df['YrSold'].unique()) == 1:
    axes = [axes]

# Plot data for each year
for i, year in enumerate(merged_df['YrSold'].unique()):
    ax = axes[i]

    # Filter data for the current year
    year_data = merged_df[merged_df['YrSold'] == year]

    ax.plot(year_data['MoSold'], year_data['SalePrice'], label='Max Sale Price', marker='o')
    
    ax.axhline(y=year_data['AvgSalePrice'].iloc[0], color='r', linestyle='--', label=f'Avg Sale Price ({year})')

    ax.set_title(f"Year {year} - Max Sale Price and Avg Sale Price by Month")
    ax.set_xlabel('Month')
    ax.set_ylabel('Sale Price')
    ax.grid(True, which='both', linestyle='--', color='gray')
    ax.legend()

plt.tight_layout()
plt.show()
```

How other factors from the Dataset affects the Sale Price and its detail description

```python
import statsmodels.api as sm

X = sm.add_constant(DATA_SET_cleaned[contnious_data_list].drop(columns=['SalePrice']))
y = DATA_SET_cleaned['SalePrice']

print(X.isnull().sum())  # Check for missing values in X
print(y.isnull().sum())  # Check for missing values in y

# If missing values are present, handle them (e.g., drop rows or fill with mean)
X = X.dropna()  # Drop rows with missing values in X
y = y[X.index]  # Align y with X after dropping rows

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Print the summary to see p-values and coefficients
model.summary()
```

Conclusion:

- Significant predictors: Variables like OverallQual, LotArea, BsmtFinSF1, GrLivArea, and YearBuilt seem to significantly influence SalePrice.
- Non-significant predictors: Variables like MasVnrArea, GarageYrBlt, and FullBath have p-values greater than 0.05, suggesting they don't significantly affect SalePrice in this model.
